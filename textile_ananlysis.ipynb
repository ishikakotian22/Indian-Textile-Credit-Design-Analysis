```python
import pandas as pd
from sqlalchemy import create_engine

engine = create_engine("postgresql+psycopg2://postgres:admin123@localhost:5432/Textile_credit_analysis")

df = pd.read_sql("SELECT * from textile_designs", engine)

df.head()
```

```python
print("Rows and Columns:", df.shape)
print("\nColumn Names:")
print(df.columns)

print("\nData Types:")
print(df.dtypes)
```

```python
no_credit = df[df["credit_given"] == "No"]

print("Brands not giving credit:")
print(no_credit[["brand_name", "brand_country", "pattern", "price"]])
```

```python
pattern_count = df["pattern"].value_counts()

print("Most reused patterns:")
print(pattern_count)
```

```python
markup_summary = df.groupby("price_markup_level")["price"].mean().sort_values()

print("Average price by markup level:")
print(markup_summary)
```

```python
country_usage = df["brand_country"].value_counts()

print("Country-wise brand adoption:")
print(country_usage)
```

```python
df["rating"] = pd.to_numeric(df["rating"], errors="coerce")
```

```python
top_patterns = (
    df.groupby("pattern")["rating"]
    .mean()
    .sort_values(ascending=False)
)

print("Highest rated patterns:")
print(top_patterns)
```
